```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

Обучение decision tree на 2х самых коррелирующих признаках

```python
X = df[['Total phenols', 'Flavanoids']]
Y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

clf = DecisionTreeClassifier(criterion='gini')
clf.fit(X_train, y_train)

predict=clf.predict(X_test)

pd.DataFrame(predict,y_test)
```

```python
print('f1-метрика= ', metrics.f1_score(y_test, predict, average= None))
print('доля правильных ответов ', metrics.accuracy_score(y_test, predict))
print('точность', metrics.precision_score(y_test, predict, average= None))
print('полнота', metrics.recall_score(y_test, predict, average= None))
```

```python
metrics. confusion_matrix (y_test, predict)
```

Обучение decision tree на 2х самых некоррелирующих признаках

```python
X = df[['Ash', 'Proanthocyanins']]
Y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

clf = DecisionTreeClassifier(criterion='gini')
clf.fit(X_train, y_train)

predict=clf.predict(X_test)

pd.DataFrame(predict,y_test)
```

```python
print('f1-метрика= ', metrics.f1_score(y_test, predict, average= None))
print('доля правильных ответов ', metrics.accuracy_score(y_test, predict))
print('точность', metrics.precision_score(y_test, predict, average= None))
print('полнота', metrics.recall_score(y_test, predict, average= None))
```

```python
metrics. confusion_matrix (y_test, predict)
```

Обучение случайного леса на всех признаках

```python
x = df.iloc[:,1:]
y = df['Class']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

RandomForest = RandomForestClassifier(n_estimators=10)
model=RandomForest.fit(x_train,y_train)
predict=RandomForest.predict(x_test)

pd.DataFrame(predict,y_test)
```

Важность признаков.

```python
pd.Series(model.feature_importances_, model.feature_names_in_,).sort_values().plot.barh()
```

```python
print('f1-метрика= ', metrics.f1_score(y_test, predict, average= None))
print('доля правильных ответов ', metrics.accuracy_score(y_test, predict))
print('точность', metrics.precision_score(y_test, predict, average= None))
print('полнота', metrics.recall_score(y_test, predict, average= None))
```

Построение ROC-кривых

```python
plt.figure(figsize=(10, 7))
for i in range(3):
    y_pred_proba = RandomForest.predict_proba(x_test)[::,i]
    fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba, pos_label=i+1)

    plt.subplot(1, 3, i+1)
    plt.plot(fpr, tpr)
    plt.title('{0} тип вина'.format(str(i+1)))
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')

plt.tight_layout()
plt.show()
```
