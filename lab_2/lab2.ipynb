```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

ÐžÐ±ÑƒÑ‡ÐµÐ½Ð¸Ðµ decision tree Ð½Ð° 2Ñ… ÑÐ°Ð¼Ñ‹Ñ… ÐºÐ¾Ñ€Ñ€ÐµÐ»Ð¸Ñ€ÑƒÑŽÑ‰Ð¸Ñ… Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ°Ñ…

```python
X = df[['Total phenols', 'Flavanoids']]
Y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

clf = DecisionTreeClassifier(criterion='gini')
clf.fit(X_train, y_train)

predict=clf.predict(X_test)

pd.DataFrame(predict,y_test)
```

```python
print('f1-Ð¼ÐµÑ‚Ñ€Ð¸ÐºÐ°= ', metrics.f1_score(y_test, predict, average= None))
print('Ð´Ð¾Ð»Ñ Ð¿Ñ€Ð°Ð²Ð¸Ð»ÑŒÐ½Ñ‹Ñ… Ð¾Ñ‚Ð²ÐµÑ‚Ð¾Ð² ', metrics.accuracy_score(y_test, predict))
print('Ñ‚Ð¾Ñ‡Ð½Ð¾ÑÑ‚ÑŒ', metrics.precision_score(y_test, predict, average= None))
print('Ð¿Ð¾Ð»Ð½Ð¾Ñ‚Ð°', metrics.recall_score(y_test, predict, average= None))
```

```python
x = df.iloc[:,1:]
y = df['Class']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

RandomForest = RandomForestClassifier(n_estimators=10)
model=RandomForest.fit(x_train,y_train)
predict=RandomForest.predict(x_test)

pd.DataFrame(predict,y_test)
```

```python
pd.Series(model.feature_importances_, model.feature_names_in_,).sort_values().plot.barh()
```

```python
print('f1-Ð¼ÐµÑ‚Ñ€Ð¸ÐºÐ°= ', metrics.f1_score(y_test, predict, average= None))
print('Ð´Ð¾Ð»Ñ Ð¿Ñ€Ð°Ð²Ð¸Ð»ÑŒÐ½Ñ‹Ñ… Ð¾Ñ‚Ð²ÐµÑ‚Ð¾Ð² ', metrics.accuracy_score(y_test, predict))
print('Ñ‚Ð¾Ñ‡Ð½Ð¾ÑÑ‚ÑŒ', metrics.precision_score(y_test, predict, average= None))
print('Ð¿Ð¾Ð»Ð½Ð¾Ñ‚Ð°', metrics.recall_score(y_test, predict, average= None))
```

ÐŸÐ¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ðµ ROC-ÐºÑ€Ð¸Ð²Ñ‹Ñ…

```python
plt.figure(figsize=(10, 7))
for i in range(3):
    y_pred_proba = RandomForest.predict_proba(x_test)[::,i]
    fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba, pos_label=i+1)

    plt.subplot(1, 3, i+1)
    plt.plot(fpr, tpr)
    plt.title('{0} Ñ‚Ð¸Ð¿ Ð²Ð¸Ð½Ð°'.format(str(i+1)))
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')

plt.tight_layout()
plt.show()
```
